# **EXPLORATORY DATA ANALYSIS (EDA)**
-----------------------------------------------------------------------------------------

**Topic** : What factors influence whether a film can be said to be successful?

```python
# Get Statistical Dataset

df.describe()
```

**1. Does Budget Affect Movie Success?**

```python
df.groupby('budget')['gross'].mean().sort_values(ascending=False)
```

```python
sns.relplot(x = 'budget', y = 'gross', estimator = 'mean', data = df, kind = 'line')
plt.title('Budget VS Gross Revenue')
plt.show()
```

**Insight & result :**

**Insights & Results:**

The analysis shows a positive relationship between budget and gross revenue, indicating that the higher a film's budget, the more likely it is to be successful. However, the wide spread of gross revenue across similar budget levels suggests that budget alone does not guarantee success. This suggests that while budget plays a supporting role, a film's success is ultimately influenced by additional factors such as audience engagement, genre appeal, and market reception.

**2. Which Genres Are Most Profitable?**

```python
df.groupby('genre')['gross'].mean().sort_values(ascending=False)
```

```python
sns.catplot(x='genre', y='gross', data=df, kind='bar', height = 5, aspect = 3)
plt.title("Box Plot of Gross Revenue by Genre")
plt.show()
```

```python
sns.relplot(x='genre', y='gross', estimator='sum', kind='line', data=df, height = 5, aspect = 3)
plt.title('Gross Revenue vs Genre')
plt.show()
```

**Results & Insigt :**

Action and Adventure dominate total industry revenue, indicating strong market demand. However, average gross per film varies significantly across genres, with Animation and Action showing high potential but also high variability. This suggests that genre selection influences revenue potential, but does not guarantee consistent success at the individual movie level

**3. Does Audience Rating Influence Revenue?**

```python
rating_gross = df.groupby('rating')['gross'].mean().sort_values(ascending=False)
rating_gross
```

```python
rating_gross = df.groupby('rating')['gross'].mean().sort_values(ascending=False)

rating_gross.plot(kind='bar')
plt.title("Mean Gross by Content Rating")
plt.show()
```

**Results & Insight :**

Movies with broader audience ratings such as TV-PG, G,and PG-13 achieve significantly higher average gross revenue. This indicates that audience reach plays a crucial role in financial success. In contrast, restrictive ratings limit market size, resulting in lower revenue potential despite possible critical acclaim

**4. How About Votes?**

```python
df.groupby('votes')['gross'].mean().sort_values(ascending = False)
```

```python
sns.relplot(x = 'votes', y = 'gross', estimator = 'mean', data = df, kind = 'line', height = 5, aspect = 3)
plt.title('Votes VS Gross Revenue')
plt.show()
```

```python
sns.scatterplot(data=df, x='votes', y='gross', alpha=0.4)
plt.title("Votes vs Gross Revenue")
plt.show()
```

**Result & Insight :**

Based on the analysis of the relationship between vote count and gross revenue, it was found that popular films tend to have a positive relationship with revenue. However, this relationship is weak and shows high variability. Films with a high vote count do not always generate high revenue, indicating that a film's success is not determined by popularity alone. Thus, votes play a supporting role, while other factors such as budget, genre, and distribution strategy are thought to have a more significant influence.

**5. Does the score influence the success of the movie?**

```python
df.groupby('score')['gross'].mean().sort_values(ascending= False)
```

```python
sns.catplot(x='score', y='gross', kind = 'bar', data =df)
plt.title('Score VS Gross Revenue')
plt.show()
```

```python
sns.relplot(x='score', y='gross', estimator='mean', kind='line', data=df)
plt.title('Score VS Gross Revenue')
plt.show()
```

**Result & Insight :**

Analysis of the relationship between score and gross revenue shows a strong positive, non-linear relationship. The higher the score given by audiences, the higher the likelihood of a film's success. Films with scores below 6 tend to generate low revenue, while revenue increases significantly once the score exceeds around 7. This suggests that film quality plays a major role in driving financial success, although the variation in revenue at higher scores suggests that quality is not the sole determinant.

**6. What is the ideal runtime for a movie?**

```python
sns.histplot(df['runtime'], bins=30)
plt.title("Movie Runtime Distribution")
plt.show()
```

```python
df.groupby('runtime')['gross'].mean().sort_values(ascending=False)
```

```python
sns.scatterplot(data=df, x='runtime', y='gross', alpha=0.4)
plt.title("Runtime vs Gross")
plt.show()
```

```python
sns.relplot(x='runtime', y = 'gross', estimator ='mean', kind = 'line', data=df )
plt.show()
```

**Result & Insight :**

Analysis of the relationship between runtime and gross revenue indicates an optimal duration range, around 170â€“200 minutes, during which films have the greatest chance of achieving high revenue. Films that are too short and too long tend to generate lower revenue. However, the large variation in revenue within the optimal duration range suggests that runtime is not the sole determinant of success, but rather a supporting factor that needs to be analyzed alongside other variables.

## **CORRELATION MATRIX**

```python
corr = df[['budget','gross','votes','score','runtime']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

**Results Correlation Matrix :**

Based on the correlation analysis, budget is the factor with the strongest relationship with gross revenue (r = 0.74), thus it can be considered as the main factor of financial success of the film. Furthermore, the number of votes also shows a fairly strong relationship (r = 0.63), which reflects the role of popularity and audience engagement. In contrast, the score and runtime have a relatively weak correlation with gross, although the score shows a moderate relationship with votes. This indicates that the quality of the film is more indirectly influenced through increasing popularity, rather than directly on revenue.

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
plt.figure(figsize=(12, 8))   # (lebar, tinggi)

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

# **CONCLUSIONS :**

Based on the results of the Exploratory Data Analysis, film success, as measured by gross revenue, is influenced by a combination of factors. Budget demonstrates the strongest relationship with revenue, making it the primary factor increasing the potential for financial success. Film quality, as reflected in the film's score, has an indirect positive influence through increased audience popularity (votes), with the effect becoming more significant after passing a certain threshold. Meanwhile, genre, age rating, and runtime act as supporting factors that influence market reach and revenue characteristics, but do not directly determine a film's success.
